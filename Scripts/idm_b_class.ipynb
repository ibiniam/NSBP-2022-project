```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
from classy import Class
```

```python
parameters_1 = {'output':'tCl, pCl, lCl, mPk',
                'lensing':'yes',
                'h':0.7,
                'omega_b':0.03,
                'omega_cdm':0.12,
                'f_idm':0.2,
                'm_idm':3e8,
                'cross_idm_b':8e-40,
                'n_index_idm_b':-4
              }

parameters_2 = {'output':'tCl, pCl, lCl, mPk',
                'lensing':'yes',
                'h':0.7,
                'omega_b':0.03,
                'omega_cdm':0.12
              }

idmb = Class()
idmb.set(parameters_1)
#idmb.set(parameters_2)
idmb.compute()
```

```python
cls = idmb.lensed_cl(2500)
print(cls)

ll = cls['ell'][2:]
clTT = cls['tt'][2:] 
clEE = cls['ee'][2:] 
clPP = cls['pp'][2:]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex = 'col', figsize = (10,10))
ax1.set_xscale('log'); ax1.set_ylabel('$\ C_l^{TT} l(l+1)/2\pi $'); ax1.plot(ll, clTT*ll*(ll+1)/2./np.pi)
ax2.set_ylabel('$\ C_l^{EE} l(l+1)/2\pi $'); ax2.plot(ll, clEE*ll*(ll+1)/2./np.pi)
ax3.set_xlabel('$\l$'); ax3.set_ylabel('$\ C_l^{PP} l^2(l+1)^2/2\pi $'); ax3.plot(ll, clPP*((ll*(ll+1))**2)/2./np.pi)

plt.show()
```
